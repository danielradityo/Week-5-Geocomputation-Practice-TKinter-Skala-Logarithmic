## Skala Logaritmik

```python
# Import Library Tkinter dan math
from tkinter import *
import math

# Membuat variabel
## Posisi Koordinat
PosXa = 100 ## Nilai posisi X (PosXa) dengan nilai 100
PosYa = 100 ## Nilai posisi Y (PosYa) dengan nilai 100
lebar = 400 ## Lebar dari skala logaritmik dengan nila 400

# Proses
## Min Max Data
mindatax = 1.4 ## Menginisiasi variabel mindatax (nilai minimal data dari x) dengan nilai 1.4
maxdatax = 80 ## Menginisiasi variabel maxdatax (nilai maximal data dari x) dengan nilai 80

## Min Max Cycle
mincycle = math.floor(math.log10(mindatax)) ## Menginisiasi variabel mincycle (nilai minimal log cycle)
maxcycle = math.ceil(math.log10(maxdatax)) ## Menginisiasi varibel maxcycle (nilia maksimal logcycle)
cycle = maxcycle - mincycle ## Menghitung banyaknya log cycle

## Skala X
skalax = lebar / cycle ## Menghitung penentuan skala X



# Pembuatan Skala Logaritmik
## Main Window
root = Tk()
root.title('Skala Logaritmik')
root.geometry('700x700')

## Canvas
canvas = Canvas(root, height = 700, width = 700, bg = 'white')

## Mebuat garis mendatar
line = canvas.create_line(PosXa, PosYa, lebar+PosXa, PosYa) ## Memindah posisi awal (0,0) menjadi posisi (PosXa, PosYa)
#linerel(lebar,0); // Menggambar garis realatif dengan lebar dari setelah di move to menuju Relatif (lebar, 0)

## Penentuan Titik X (menggunakan while Loop)
i = 0
while i <= cycle: ## Perulangan dengan nilai i awal adalah 0 sampai kurang dari sama dengan cycle
    posx = PosXa + (i * skalax) ## Menginisiasi nilai posx dengan data PosXa(titik awal) ditambah dengan variabel i kemudian dikali skalax
    line = canvas.create_line(posx, PosYa-5, posx, PosYa+5) ## untuk membuat garis-garis turus kebawa
    xint = math.pow (10, i) ## menginisiasi nilai xint sebagai nilai i berpangkat 10 sebagai angka logaritmiknya
    text = canvas.create_text(posx, PosYa-15, text=xint) ## untuk menampilkan angka logaritmiknya
    i += 1 ## Membatasi Looping dengan menambah 1 hingga sesuai dengan persyaratan looping
    
    ## Ada if dalam loopingan While untuk membuat Minor Unitnya
    if (i < cycle + 1): ## pengecekan data cycle hingga benar baru program if berjalan
        minor = 1 ## Menginisiasi variabel minor dan harus 1 karena tidak ada log 0
        while minor < 10: ## perulangan nilai minor dengan awal nilai minor = 1 sampai kurang dari 10
            minorint = (math.log10(minor)) ## menetapkan bilangan konstan minor menggunakan logaritmik
            posminor = posx + (minorint * skalax) ## menentukan nilai posminor
            line = canvas.create_line(posminor,PosYa-5,posminor,PosYa) ## menampilkan garis dari data posminor, PosYa - 5, posminor, PosYa
            minor += 1 ## Membatasi Looping dengan menambah 1 hingga sesuai dengan persyaratan looping

canvas.pack()
root.mainloop()
```

## Codingan C++ (Jangan di run)

```python
//TUGAS GEOKOMPUTASI LOGARITMIK

#include <iostream> // digunakan perintah include standart
#include <cmath> // digunakan perintah statement matematika standart
#include <graphics.h> // digunakan perintah statement untuk menampilkan graphic
#include <fstream> //untuk menampilkan text berupa string dari/ke file

using namespace std; // kepala proram dan penggunaan iostream

int main () // baris deklarasi fungsi awal
{
    initwindow (1000, 600, "Logaritmik"); //statement untuk membuat window utama graphic dengan judul "Logaritmik"
    int PosXa = 100; // Nilai posisi X (PosXa) dengan nilai 100
    int PosYa = 100; // Nilai posisi Y (PosYa) dengan nilai 100
    float lebar = 400; // Lebar dari skala logaritmik dengan nila 400

    moveto(PosXa, PosYa); // Memindah posisi awal (0,0) menjadi posisi (PosXa, PosYa)
    linerel(lebar,0); // Menggambar garis realatif dengan lebar dari setelah di move to menuju Relatif (lebar, 0)

    float mindatax = 1.4; // Menginisiasi variabel mindatax (nilai minimal data dari x) dengan nilai 1.4
    float maxdatax = 80; // Menginisiasi variabel maxdatax (nilai maximal data dari x) dengan nilai 80

    float mincycle=floor(log10(mindatax)); // Menginisiasi variabel mincycle (nilai minimal log cycle)
    float maxcycle=ceil(log10(maxdatax)); // Menginisiasi varibel maxcycle (nilia maksimal logcycle)
    float cycle=maxcycle-mincycle; // Menghitung banyaknya log cycle

    float skalax = lebar / cycle; // Menghitung penentuan skala X

    int i; // memasukan nilai varibel i berupa integer
    for (i=0; i <= cycle; i++) // Perulangan dengan nilai i awal adalah 0 sampai kurang dari sama dengan cycle
    {
        float posx = PosXa + (i * skalax); // Menginisiasi nilai posx dengan data PosXa(titik awal) ditambah dengan variabel i kemudian dikali skalax
        line (posx,PosYa-5,posx, PosYa+5); // untuk membuat garis-garis turus kebawah

        float xint = pow (10, i); // menginisiasi nilai xint sebagai nilai i berpangkat 10
        char textx [5]; // Menginisiasi karakter pada textx dengan nilai 5
        snprintf(textx,5,"%f",xint); // Mengubah data angka menjadi data kata

        float lebarteks = textwidth(textx); // menetapkan lebar text dalam variabel lebarteks
        float tinggiteks = textheight(textx); // menetapkan tinggi text dalam variabel tinggiteks
        float postextxx = posx-lebarteks/2; //menentukan nilai postextxx
        float postextxy = PosYa-2*tinggiteks; //menentukan nilai postextxy
        outtextxy(postextxx, postextxy, textx); // Fungsi untuk mengeluarkan teks.

        int minor; //menginisiasi variabel minor
        if (i < cycle) // pengecekan data cycle hingga benar baru program if berjalan
        {
            for (minor=1; minor<=10; minor++) // perulangan nilai minor dengan awal nilai minor = 1 sampai kurang dari 10
            {
                float minorint(log10(minor)); // menetapkan bilangan konstan minor menggunakan logaritmik
                float posminor=posx+(minorint*skalax); // menentukan nilai posminor
                line(posminor,PosYa-5,posminor,PosYa); // menampilkan garis dari data posminor, PosYa - 5, posminor, PosYa
            }
        }
    }

    while (!kbhit());//untuk menunggu user untuk memencet tombol apapun.
    closegraph();
    return 0;
}
```
